```python
from selenium import webdriver
import requests
import time
import bs4 as bs

url = 'https://www.magicbricks.com/property-for-rent/residential-real-estate?bedroom=2&proptype=Multistorey-Apartment,Builder-Floor-Apartment,Penthouse,Studio-Apartment,Service-Apartment&cityName=Bangalore&BudgetMax=20000'
# 'https://www.magicbricks.com/'
driver = webdriver.Chrome(r'C:\Users\madhu\Downloads\chromedriver.exe')
driver.get(url)
driver.implicitly_wait(100)
# d = driver.find_elements_by_xpath("/input[contains(@id,'fromdate')]")
# d.send_keys('Banglore')
ScrollNumber = 5
for i in range(1,ScrollNumber):
    driver.execute_script("window.scrollTo(0,document.body.scrollHeight)")
    time.sleep(3)
#     driver.execute_script("window.scrollTop(1,5)")
#     time.sleep(3)

with open('DS.html', 'w', encoding='utf-8') as f:
    f.write(driver.page_source)

data = open('DS.html','r')
soup = bs.BeautifulSoup(data, 'html.parser')
print(soup)    

driver.close()
```
